## If-Else

```python
highest_income = 50000
income= 40000
```

```python
income< highest_income
```

```python
if income< highest_income:
    print("Income is smaller than highest income") # It computes here if condition above is right
    print("We can right here whatever we want")
```

```python
highest_income = 30000
income= 40000
```

```python
income< highest_income
```

```python
if income< highest_income:
    print("Income is smaller than highest income") # It doesn't compute here if condition above is false
```

```python
if income< highest_income:
    print("Income is smaller than highest income")
else:
    print("Here is else statement")
    print("Income is not smaller than highest income") # We can use else statement if condition is false
```

## Grade System User Interaction Example

```python
note = float(input("Please enter your note:"))

if note >= 90:
    print("AA")
elif note >= 85:
    print("BA")
elif note >= 90:
    print("BA")
elif note >= 80:
    print("BB")
elif note >= 75:
    print("CB")
elif note >= 70:
    print("CC")
elif note >= 65:
    print("DC")
elif note >= 60:
    print("DD")
else:
    print("You did not pass the exam!")
```

## Nested If

```python
grade5 = 80
```

```python
if grade5>50:
    if grade5>=75:
        print("Grade is between 75 and 100. Very Good!")
    else:
        if grade5>60:
            print("Grade is between 60 and 75. Good!")
        else:
            print("Grade is between 50 and 60. Enough!")
elif grade5==50:
    print("Grade is 50!")
else:
    print("Student can't pass the exam!")
```
